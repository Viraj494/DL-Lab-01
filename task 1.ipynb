```python
import matplotlib.pyplot as plt

# Create random arrays
exp_data = np.random.exponential(scale=1, size=100000)
uniform_data = np.random.rand(100000)
normal_data = np.random.randn(100000)

# Plot histograms
plt.hist(exp_data, density=True, bins=100, histtype="step", color="red", label="Exponential")
plt.hist(uniform_data, density=True, bins=100, histtype="step", color="blue", label="Uniform")
plt.hist(normal_data, density=True, bins=100, histtype="step", color="green", label="Normal")
plt.axis([0, 5, 0, 1.1])  # Adjusted axis for better view
plt.legend(loc="upper right")
plt.title("Random distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 + Y**2

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='viridis')

ax.set_title("3D plot of Z = X^2 + Y^2")
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Data provided by the user
data = {
    "Name": ["Bulbasaur", "Ivysaur", "Venusaur", "Charmander", "Charmeleon"],
    "Type 1": ["Grass", "Grass", "Grass", "Fire", "Fire"],
    "Type 2": ["Poison", "Poison", "Poison", np.nan, np.nan],
    "Total": [318, 405, 525, 309, 405],
    "HP": [45, 60, 80, 39, 58],
    "Attack": [49, 62, 82, 52, 64],
    "Defense": [49, 63, 83, 43, 58],
    "Sp. Atk": [65, 80, 100, 60, 80],
    "Sp. Def": [65, 80, 100, 50, 65],
    "Speed": [45, 60, 80, 65, 80],
    "Stage": [1, 2, 3, 1, 2],
    "Legendary": [False, False, False, False, False]
}

# Convert data to DataFrame
df = pd.DataFrame(data)

# Selecting the relevant features
features = df[["HP", "Attack", "Defense", "Sp. Atk", "Sp. Def", "Speed"]]

# Calculating Pearson correlation coefficient
pearson_corr = features.corr(method='pearson')

# Calculating Spearman rank correlation
spearman_corr = features.corr(method='spearman')

# Creating heatmaps
plt.figure(figsize=(12, 6))

# Pearson correlation heatmap
plt.subplot(1, 2, 1)
sns.heatmap(pearson_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Pearson Correlation')

# Spearman rank correlation heatmap
plt.subplot(1, 2, 2)
sns.heatmap(spearman_corr, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Spearman Rank Correlation')

plt.tight_layout()
plt.show()
```
